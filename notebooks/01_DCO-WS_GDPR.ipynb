## Parsing high level table

```python
driver = webdriver.Edge('../srcs/msedgedriver.exe')
driver.get('https://www.enforcementtracker.com/')
time.sleep(4)

# Select dropdown by name
x = driver.find_element_by_name('penalties_length')
drop = Select(x)
  
# Select by value
drop.select_by_value("100")
time.sleep(4)
```

```python
table = driver.find_element_by_xpath('//*[@id="penalties_wrapper"]')
table_html = table.get_attribute('innerHTML')
```

```python
df = pd.read_html(table_html)[0]
df.dropna(how='all', inplace=True)
df
```

```python
# number of pages to parse
page_no = 10

for i in range(1,page_no):
    print(i)
    time.sleep(20)
    try:
        driver.find_element_by_id('penalties_next').click()
    except:
        print("Next Click failed. Trying it again.")
        time.sleep(20)
        driver.find_element_by_id('penalties_next').click()
    time.sleep(4)
    table = driver.find_element_by_xpath('//*[@id="penalties_wrapper"]')
    table_html = table.get_attribute('innerHTML')
    df2 = pd.read_html(table_html)[0]
    df2.dropna(how='all', inplace=True)
    df = df.append(df2)
```

## Parsing single case/fine details

**Attention:** This process takes a few hours, because 914 pages have to be opened and parsed in sequence.

```python
# Import required module
import time
from selenium import webdriver
  
# Import Select class
from selenium.webdriver.support.ui import Select
```

```python
df_length = len(df)
```

```python
#df_length = 3

# For each entry in the dataframe open the respective case website and parse the details
for i in range(0, df_length):
    ETid_no = i+1
    driver = webdriver.Edge('../srcs/msedgedriver.exe')
    try:
        driver.get(f'https://www.enforcementtracker.com/ETid-{ETid_no}')
        time.sleep(3)
    except:
        print("Opening source failed. Trying it again.")
        driver.get(f'https://www.enforcementtracker.com/ETid-{ETid_no}')
        time.sleep(3)  
    
    table = driver.find_element_by_xpath('//*[@id="penalties"]/tbody/tr[1]')
    table_html = table.get_attribute('innerHTML')
    
    split_text = table_html.split(">")
    split_text = [text for text in split_text if text.startswith('<') == False]
    split_text
    
    authority = split_text[2][:-4]
    sector = split_text[6][:-4]
    summary = split_text[9][:-4]
    
    print(ETid_no)
    print(authority)
    print(sector)
    print(summary)

    df.loc[f'ETid-{ETid_no}','Authority'] = authority
    df.loc[f'ETid-{ETid_no}','Sector'] = sector
    df.loc[f'ETid-{ETid_no}','Summary'] = summary

    driver.close()
```

```python
df
```

```python
df.reset_index(inplace=True)
df
```
